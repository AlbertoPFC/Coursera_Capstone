```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
df_toronto = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788')
df_toronto
```

```python
print(f'Número total de tablas: {len(df_toronto)}')
```

```python
df_codto = df_toronto[0]
df_codto.head(12)
```

```python
df_codto.columns = ['CódigoPostal','Municipio','Vecindario']
df_codto.head(12)
```

```python
df_codto2 = df_codto[df_codto['Municipio']!= 'Not assigned']
df_codto2 = df_codto2.reset_index(drop=True)
df_codto2.head(12)
```

```python
df_codto3 = df_codto2.groupby(['CódigoPostal','Municipio'])['Vecindario'].agg(lambda x: tuple(x)).reset_index()
df_codto3.head(12)
```
